# Coin Change Infinite Coins

```python
# Dynamic Programming Python implementation of Coin
# Change problem


def count(coins, target):

    # dp[i] will be storing the number of solutions for
    # value i. We need target+1 rows as the dp is constructed
    # in bottom up manner using the base case (target = 0)
    # Initialize all table values as 0
    coins = list(set(coins))
    n = len(coins)
    dp = [False for k in range(target+1)]

    # Base case (If given value is 0)
    dp[0] = True

    # Pick all coins one by one and update the dp[] values
    # after the index greater than or equal to the value of the
    # picked coin
    print(f"coins={coins}, n={n}, target={target}, dp={dp}\n")
    for i in range(0, n):
        for j in range(coins[i], target+1):
            dp[j] |= dp[j-coins[i]]
            print(f"coins[i={i}]={coins[i]}, j={j}, dp={dp}")
    # print(f"dp = {dp}")

    return dp[target]


def printResults(coins, target):
    x = count(coins, target)
    print()
    print(f"Is it possible to achieve the target {target}: {x}")

# Driver program to test above function
coins = [3, 5]
# printResults(coins, 6)
# print("\n*********************************************\n")
# printResults(coins, 7)
# print("\n*********************************************\n")
# printResults(coins, 8)
# print("\n*********************************************\n")
printResults(coins, 18)
print("\n*********************************************\n")


# This code is contributed by Afzal Ansari
```
